### 03. Combining Dataframes

The best way to combine the df_orders_prods_combined and the df_prods dataframes is to use the merge function. The two main reasons to combine them this way is that the two dataframes have different shapes, and they both share a common key in the product_id column in both dataframes.

```python
# Testing merge without overwriting
pd.merge(df_orders_prods_combined, df_prods, on = 'product_id', indicator = True) 
```

When testing the merge, the _merge column already exists in the df_orders_prods_combined dataframe, so the column needs to be dropped in order to merge the two dataframes.

```python
# Dropping _merge column in order to merge dataframes
df_orders_prods_combined = df_orders_prods_combined.drop(['_merge'], axis = 1)

# Checking dataframe
df_orders_prods_combined.head()
```

```python
# Testing merge without overwriting again
pd.merge(df_orders_prods_combined, df_prods, on = 'product_id', indicator = True) 
```

```python
# Merging dataframes into new dataframe
df_ords_prods_merged = df_orders_prods_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# Checking df_ords_prods_merged dataframe
df_ords_prods_merged.head()
```

### 02. User-Defined Functions

```python
df_ords_prods_merge.head()
```

```python
# Creating subset of df_ords_prods_merge
df = df_ords_prods_merge[:1000000]

# Price label function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'

# Running price_label function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Checking values of new column
df['price_range'].value_counts(dropna = False)
```

### 03. If-Statements with the loc() Function

```python
# Using loc() function in if-statements
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Using loc() function on df_ords_prods_merge
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### 04. If-Statement with For-Loops

```python
# Finding busiest day
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating for loop
result = []

for value in df_ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

### 05. Task Exercises

```python
# Creating busiest_days colunmn that labels the two busiest days as busiest and the two least busy as least busy
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

The two busiest days are Saturday (0) and Sunday (1). The two least busy days are Tuesday (3) and Wednesday (4)

```python
result = []

# Creating for loop to check for busiest days
for value in df_ords_prods_merge['order_day_of_week']:
    if (value == 0) or (value == 1):
        result.append('Busiest days')
    elif (value == 3) or (value == 4):
        result.append('Least busy')
    else:
        result.append('Regularly busy')

# Creating new column named busiest_days
df_ords_prods_merge['busiest_days'] = result
```

```python
# Checking values of busiest_days column
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

Busiest Days: Saturday + Sunday (6204182 + 5660230) = 11864412 which is the Busiest days count
Least Busy Days: Tuesday + Wednesday (3840534 + 3783802) = 7624336 which is the Least busy days count

```python
# Identifying the busiest times of day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

For identifying the busiest times of day there will be 3 labels:
The top 8 order_hour_of_day frequencies will be labeled as Most orders (9, 10, 11, 12, 13, 14, 15, 16).
The bottom 8 order_hour_of_day frequencies will be labeled as Fewest orders (0, 1, 2, 3, 4, 5, 6, 23).
Everything else will be labeled as Average orders

```python
result = []

# Creating for loop to get busiest hours of day
for value in df_ords_prods_merge['order_hour_of_day']:
    if value in [9, 10, 11, 12, 13, 14, 15, 16]:
        result.append('Most orders')
    elif value in [0, 1, 2, 3, 4, 5, 6, 23]:
        result.append('Fewest orders')
    else: 
        result.append('Average orders')

result
```

```python
# Creating new column named busiest_period_of_day
df_ords_prods_merge['busiest_period_of_day'] = result
```
